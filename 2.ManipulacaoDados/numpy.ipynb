#### Arrays X Listas

```python
a = [1, 2, 3, 4]
b = [4, 8, 3, 1]
```

```python
# Qual o resultado de a + b ?
# Qual o resultado de a - b ?
# Qual o resultado de a*3 ?
print(a+b)
print(a*3)
```

```python
a_np = np.array([1, 2, 3, 4])
b_np = np.array([4, 8, 3, 1])
```

```python
# Qual o resultado de a_np + b_np ?
# Qual o resultado de a_np - b_np ?
# Qual o resultado de a_np*3 ?
print(a_np + b_np)
print(a_np - b_np)
print(a_np*3)
```

```python
print(type(a), type(a_np))
```

#### Matrizes (arrays n-dimensional)

```python
M = np.array([[1,2], [0,2], [2,1]])
print(M)
```

```python
M.shape
```

```python
# Primeira Linha
print(M[0])
# Primeira coluna
print(M[:,0])
# Subset da matriz
print(M[:2, :2])
```

```python
# Atribui valor a um elemento específico
M[0,0] = 10
print(M)
```

3. Operações de Algebra Linear

```python
x = np.array([1, 2, 3, 4])
w = np.array([5, 6, 7, 8])
```

Combinação linear: $\ x_1*w_1 + x_2*w_2 + x_3*w_3+ . . . + x_m*w_m$

```python
sum(x*w)
```

```python
np.dot(x,w)
```

Multiplicação de Matrizes

```python
a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])
```

```python
a@b
```

```python
np.dot(a, b)
```

Matriz Inversa: $\ A^{-1} . A = I$

```python
np.linalg.inv(a)
```
